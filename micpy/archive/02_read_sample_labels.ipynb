```python
from pathlib import Path
import cv2
from config import original_images, sample_labels, sample_splitted_images
from matplotlib import pyplot as plt
import numpy as np
```

```python
images = []
labels = []

for img in sample_splitted_images[:3]:
    if "6754_46579_RSFA_50x0052" in Path(img).name:
        images.append(cv2.imread(img, 0))

for lbl in sample_labels:
    if "6754_46579_RSFA_50x0052" in Path(lbl).name:
        labels.append(cv2.imread(lbl, 0))
```

```python
fig, axis = plt.subplots(3,2 , figsize=(10,7))
fig.tight_layout()
for i in range(len(images)):
    axis[i,0].imshow(images[i], cmap="bone")
    axis[i,1].imshow(labels[i], cmap="bone")
```
