```python
from pathlib import Path
import cv2
import numpy as np
from matplotlib import pyplot as plt
from config import images, labels
```

## Find original images for each label and store them in corresponding variables

```python
names = []
images_arr = []
labels_arr = []
for img in images:
    name = Path(img).name
    stem = Path(img).stem

    for lbl in labels:
        if stem in lbl:
            names.append(name)
            images_arr.append(cv2.imread(img, 0))
            labels_arr.append(cv2.imread(lbl, 0))
```

```python
images_arr = np.array(images_arr)
labels_arr = np.array(labels_arr)
```

```python
fig, axis = plt.subplots(1,2)
print(names[4])
axis[0].imshow(images_arr[4], cmap="bone")
axis[1].imshow(labels_arr[4], cmap="bone")
```

```python
def convert_binary_labels_to_grayscale():
    for i, lbl in enumerate(labels_arr):
        lbl[lbl==1] = 255
        print(lbl)
        cv2.imwrite(f"labels/{names[i]}", lbl)
```
